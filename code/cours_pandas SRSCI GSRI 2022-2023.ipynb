## Accès indicé aux données d'un DataFrame
On peut accéder aux valeurs du DataFrame via des indices ou plages d'indice. La structure se comporte alors comme une
matrice. La cellule en haut et à gauche est de coordonnées (0,0).
Il y a différentes manières de le faire, l'utilisation de .iloc[,] constitue une des solutions les plus simples. N'oublions pas
que Shape permet d'obtenir les dimensions (lignes et colonnes) du DataFrame.

```python
#accès à la valeur située en (0,0)
print(df2.iloc[0,0])
```

```python
#valeur située en dernière ligne, première colonne
#utilisation de l'indiçage négatif
print(df2.iloc[-1,0])
```

```python
#5 premières valeurs de toutes les colonnes
#lignes => 0:5 (0 à 5 [non inclus])
#colonnes = : (toutes les colonnes)
print(df2.iloc[0:5,:])
```

```python
#5 premières valeurs des colonnes 2,3,4,5
#lignes => 0:5 (0 à 5 [non inclus])
#colonnes = 2,3,4,5
print(df2.iloc[0:5,2:6])
```

```python
#avec l'indiçage négatif, on peut facilement accéder aux 5 dernières lignes
print(df2.iloc[-5:,:])
```

```python
#5 premières lignes et colonnes 0, 1 et 4
#on a une liste d'indices en colonne
print(df2.iloc[0:5,[0,2,4]])
```

```python
#ou encore, pour exactement la même chose
#remarquez le rôle de 2 dans 0:5:2
print(df2.iloc[0:5,0:5:2])
```

# Restrictions avec les conditions - Les requêtes
Nous pouvons isoler les sous-ensembles d'observations répondant à des critères définis sur les champs. Nous utiliserons
préférentiellement la méthode .loc[,] dans ce cadre.

```python
#liste des individus présentant une douleur de type A
print(df2.loc[df2['type_douleur']=="A",:])
```

```python
print((df2['type_douleur']=="A").value_counts())
```

```python
print(df2.loc[df2['type_douleur']!="A",:])
```

```python
#pour un ensemble de valeurs de la même variable,
#nous utilisons isin()
print(df2.loc[df2['type_douleur'].isin(['A','B']),:])
```

```python
#Des opérateurs logiques permettent de combiner les conditions. 
#Nous utilisons respectivement : & pour ET, | pour OU, et ~ pour la négation.
#liste des individus présentant une douleur de type A et angine == oui
print(df2.loc[(df2['type_douleur']=="A") & (df2['angine'] == "oui"),:])
```

```python
#liste des personnes de moins de 45 ans, de sexe masculin, présentant une
#maladie cardiaque
print(df2.loc[(df2['age'] < 45) & (df2['sexe'] == "masculin") & (df2['coeur'] == "presence"),:])
```
